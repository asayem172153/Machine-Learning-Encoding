# Replace

```python
object_dict={}
count=0 #we can use random value
for column in df2.columns:
    if df2[column].dtype == 'object':
        uniqueValue=df2[column].unique()
        for value in uniqueValue:
            object_dict[value]=count
            count+=1
```

```python
for column in df2.columns:
    if df2[column].dtype =='object':
        df2[column] = df2[column].replace(object_dict)
```

```python
def labelEncoding(column):
    object_dict={}
    count=0 #we can use random value
    uniqueValue=column.unique()
    for i in uniqueValue:
        object_dict[i]=count
        count+=1
    return object_dict
```

```python
for column in df3.columns:
    if df3[column].dtype =='object':
        df3[column] = df3[column].replace(labelEncoding(df3[column]))
```
